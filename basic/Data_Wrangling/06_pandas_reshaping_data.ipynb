```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("mpg")
df.shape
```

```python
df.head()
```

```python
df.sort_values('mpg').head()
```

```python
# ascending=False -> 역순으로 정렬
# ascending=True -> 순서대로 정렬(기본값)
# descending은 사용하지 않음
df.sort_values('mpg', ascending=False).head()
```

```python
# 컬럼 이름 변경 후 df에 다시 넣어줘야 완벽하게 원본이 변경됨(또는 inplace=True 사용)
df = df.rename(columns = {'model_year': 'year'}).head()
```

```python
df
```

```python
# 인덱스별로 정렬
df.sort_index()
```

```python
# 인덱스가 없을 때, 인덱스를 새로 만들고 싶을 때 사용
df.reset_index()
```

```python
# inplace=True로 데이터프레임 변경 즉시 원본을 고침
df.drop(columns=['mpg', 'year', 'name'], inplace=True)
```
