```python
ts = pd.Series(np.random.randn(1000),
               index=pd.date_range('1/1/2000', periods=1000))
```

```python
ts
```

```python
ts.plot()
```

```python
ts = ts.cumsum()
```

```python
ts.plot()
```

```python
# 1000개의 행과 4개의 열을 랜덤으로 생성
df = pd.DataFrame(np.random.randn(1000, 4),
                  index=ts.index, columns=list('ABCD'))
```

```python
df
```

```python
# cumsum() 전 데이터: 각 시점의 독립적인 변화량을 보여줌(불규칙한 모습)
df.plot()
```

```python
df = df.cumsum()
```

```python
# cumsum() 후 데이터: 데이터의 누적 값이므로 추세가 생김
# 값이 양수면 그래프가 우상향, 음수면 우하향 -> 전체적인 흐름을 파악하기 쉬운 모습
df.plot()
```
