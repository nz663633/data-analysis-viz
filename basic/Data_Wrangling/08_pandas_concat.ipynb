```python
import pandas as pd
```

```python
s1 = pd.Series(['a', 'b'])
s1
```

```python
s2 = pd.Series(['c', 'd'])
s2
```

```python
# concat(): 둘을 하나로 합치기
# ignore_index=True:  index를 무시
pd.concat([s1, s2], ignore_index=True)
```

```python
# keys=['s1', 's2']: 합쳐진 시리즈의 바깥쪽 인덱스(Level 0) 이름을 결정
# names=['Series name', 'Row ID']: 생성된 멀티인덱스의 각 층(Level)에 이름을 부여
pd.concat([s1, s2], keys=['s1', 's2'], names=['Series name', 'Row ID'])
```

```python
pd.concat([s1, s2], axis=1)
```

```python
df1 = pd.DataFrame([['a', 1], ['b', 2]],
                   columns=['letter', 'number'])

df1
```

```python
df2 = pd.DataFrame([['c', 3], ['d', 4]],
                   columns=['letter', 'number'])

df2
```

```python
pd.concat([df1, df2])
```

```python
df3 = pd.DataFrame([['c', 3, 'cat'], ['d', 4, 'dog']],
                   columns=['letter', 'number', 'animal'])

df3
```

```python
# 형태가 다른 데이터프레임 합치기
pd.concat([df1, df3])
```

```python
# join='inner': 두 데이터프레임에서 공통된 컬럼(또는 인덱스)만 남겨서 출력
pd.concat([df1, df3], join='inner')
```

```python
df4 = pd.DataFrame([['bird', 'polly'], ['monkey', 'george']],
                   columns=['animal', 'name'])

df4
```

```python
df5 = pd.DataFrame([1], index=['a'])

df5
```

```python
df6 = pd.DataFrame([2], index=['a'])

df6
```

```python
# verify_integrity=True: 데이터를 합친 후, 중복된 인덱스가 있다면 에러 발생
# concat의 기본 방향은 axis=0(행)이므로 'verify_integrity=True'는 컬럼에 대해서 해당 X
pd.concat([df5, df6])
```
