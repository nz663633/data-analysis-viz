```python
df = pd.DataFrame(
     [[4, 7, 10],
      [5, 8, 11],
      [6, 9, 12]], 
     index=[1, 2, 3], 
     columns=['a', 'b', 'c'])

df
```

```python
# 튜플 형태의 데이터로 이중 인덱스 생성
# np.nan -> Numpy(np) 라이브러리에 있는 nan 값을 사용하겠다

import numpy as np
df = pd.DataFrame(
    {"a" : [4 ,5, 6, 6, np.nan], 
    "b" : [7, 8, np.nan, 9, 9], 
    "c" : [10, 11, 12, np.nan, 12]},    
index = pd.MultiIndex.from_tuples(
    [('d', 1), ('d', 2),('e', 2),('e', 3),('e', 4)],
    names=['n', 'v']))

df
```

```python
# 비교 연산자로 열의 값 비교하기(1)
df[df.b > 7]
```

```python
# 비교 연산자로 열의 값 비교하기(2)
df[df['c'] > 10]
```

```python
df
```

```python
# 중복된 행 제거
# keep='last' -> 중복된 데이터가 발견되었을 때
# 가장 마지막에 나온 데이터만 남기고 다 지우기
df = df.drop_duplicates(keep='last')
df
```

```python
df["a"] != 7
```

```python
# 'b'열의 값이 7이 아닌 행들만 추출(7이 포함된 행은 제외)
df[df["b"] != 7]
```

```python
# 'a'열에 5와 6이 있는가?
df.a.isin([5, 6])
```

```python
import pandas as pd
df = pd.DataFrame({'a': [1, 10, 8, 11, -1],
                'b': list('abcde'),
                'c': [1.0, 2.0, np.nan, 3.0, 4.0]})

df
```

```python
# 특정 열을 기준으로 가장 큰 값을 가진 상위 n개의 행을 추출
# 'a'열에서 가장 큰 값을 가진 상위 3개의 행 추출
df.nlargest(3, 'a')
```

```python
# 특정 열을 기준으로 가장 작은 값을 가진 n개의 행을 추출
df.nsmallest(2, 'c')
```
