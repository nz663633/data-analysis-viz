### apply(function)

````python
def smp(x):
    # 뒤에서 3번째까지의 문자를 가져오는 함수
    x = x[-3:] 
    return x
````

````python
# 각 열에서 인덱스가 1인 데이터들만 추출
df.apply(lambda x : x[1])
````

````python
# 'species'열에서 3번째 문자까지만 가져와서 출력
# df의 새로운 열을 생성해서 담기
df['species_3'] = df['species'].apply(lambda x : x[:3])
````

````python
df['species-3'] = df['species'].apply(smp)
````
