### Hexagonal Bin Plot
- Hexbin plot은 데이터가 클 때 각각의 점을 산점도(scatter plot)로 표현할 때의 단점을 보완할 수 있다.

```python
import pandas as pd
import numpy as np
import matplotlib as mpl

%matplotlib inline

# 한글폰트 사용시 그래프에서 마이너스 폰트 깨지는 문제에 대한 대처
mpl.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
# np.arange(1000): 데이터프레임의 행 번호(0~999)를 각 행의 'b'컬럼 값에 더해줌
df['b'] = df['b'] + np.arange(1000)
```

```python
df.head()
```

```python
df.tail()
```

```python
# gridsize=25: x축 방향으로 놓일 육각형의 개수를 25개로 지정
df.plot.hexbin(x='a', y='b', gridsize=25)
```

```python
# uniform(0, 3, 1000): 0~3까지 범위의 숫자 1000개 생성
df['z'] = np.random.uniform(0, 3, 1000)
```

```python
df['z']
```

```python
# C='z': 색상의 기준으로 'z'컬럼의 값으로 변경(데이터 개수 X)
# reduce_C_function=np.max: 'z'컬럼의 값들 중 최댓값을 기준으로 색을 입힘
df.plot.hexbin(x='a', y='b', C='z', 
               reduce_C_function=np.max, gridsize=20)
```
