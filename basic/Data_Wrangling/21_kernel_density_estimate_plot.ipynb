```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

%matplotlib inline

# 한글폰트 사용시 그래프에서 마이너스 폰트 깨지는 문제에 대한 대처
mpl.rcParams['axes.unicode_minus'] = False
```

```python
# 정규분포(randn)를 이루어야 머신러닝, 딥러닝에서 좋은 성능을 낼 수 있다.
# 되도록 정규분포로 만들어주는 것이 좋다.
ser = pd.Series(np.random.randn(1000))
```

```python
# y축 denstiy
ser.plot.density()
```

```python
# density()와 kde()는 같은 그래프
# 데이터의 전반적인 흐름과 패턴 파악 용이
ser.plot.kde()
```

```python
# 이산적 - 실제 데이터 빈도를 정확히 확인 가능
ser.plot.hist()
```
