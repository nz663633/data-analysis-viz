```python
import pandas as pd
import numpy as np
```

```python
# np.random.randn(10) -> 정규분포를 따르는 10개의 숫자를 1차원 배열로 생성
df = pd.DataFrame({'A': range(1,11),
                   'B': np.random.randn(10)})
df
```

```python
# assign(): 기존 데이터프레임을 수정하지 않고, 새로운 컬럼을 붙여 새로운 데이터프레임 반환하는 함수
# 'ln_A'라는 이름의 새로운 컬럼 생성하기
# 'A'컬럼의 log 연산 수행해서 ln_A 컬럼에 저장
df.assign(ln_A = lambda x : np.log(x.A)).head()
```

### apply와 assign의 차이?

* apply: 행이나 열을 따라 사용자가 정의한 함수를 반복 적용할 때 사용
1. 기존 열의 값을 수정하거나, 여러ㅓ 열의 값을 조합해 하나의 결과를 낼 때 유용
2. 복잡한 조건문이나 함수를 모든 데이터에 일일이 적용할 때 유용

* assign: 기존 데이터프레임을 건드리지 않고, 새로운 열이 추가된 새로운 데이터프레임을 반환할 때 사용
1. 새로운 열을 생성하거나 기존 열을 덮어씌우는 '구조적 작업'에 최적화
2. 메서드 체이닝(원본 데이터 수정 X, 여러 명령을 한 줄로 이어서 쓸 때 유용)

```python
# 위의 방식보다 이 방식이 더 표준
df["ln_A"] = np.log(df.A)
df
```

```python
# qcut(): 동일한 개수(빈도)를 가진 구간으로 나눌 때 사용
# qcut(컬럼, 나눌 구간의 수, 구간의 이름)
pd.qcut(df.A, 3, labels=["good", "medium", "bad"])
```

```python
# 열을 기준으로 최댓값
df.max(axis=1)
```
