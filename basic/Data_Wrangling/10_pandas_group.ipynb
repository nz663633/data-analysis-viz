```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("mpg")
df.head(3)
```

```python
# origin으로 그룹핑
# 각 그룹의 최댓값 출력
df.groupby(by="origin").max()
```

```python
# df.groupby(by="origin").size()
df["origin"].value_counts()
```

```python
# model_year, origin으로 데이터를 두 컬럼으로 분리
# cylinders 컬럼만 선택해 평균값을 계산

# model_year와 origin을 기준으로 그룹화하고,
# cylinders의 평균을 구해 데이터프레임으로 변환
pd.DataFrame(df.groupby(['model_year', 'origin'])['cylinders'].mean())
```

```python
df2 = pd.DataFrame(
    [[4, 7, 10],
     [5, 8, 11],
     [6, 9, 12]],
     index=[1, 2, 3],
     columns=['a', 'b', 'c']
)
```

```python
df2
```

```python
# row(행)가 아래로 한 칸씩 이동(shift)
df2.shift(1)
```

```python
df2
```

```python
# 특정 컬럼 'b'만 아래로 2칸 밂
df2['b'].shift(2)
```

```python
# 특정 컬럼 'a'만 위로 1칸 밂
df2['a'].shift(-1)
```

```python
# model_year의 컬럼 값들에 대해 순위를 매김
# method='dense': 밀집순위 계산
df["model_year"].rank(method='dense')
```

```python
# model_year의 데이터 갯수 중에서 해당 순위가 어느 위치에 있는지 백분위로 출력(0 ~ 1)
df["model_year"].rank(pct=True).head()
```

```python
df2
```

```python
# 각 열별 누적 합 구하기
# 각 행을 지나갈 때마다 이전 행까지의 합계를 더하며 내려옴
df2.cumsum()
```

```python
# 각 열별 누적 곱 구하기
# 아래로 내려가면서 이전 값들을 차례대로 곱함
df2.cumprod()
```
