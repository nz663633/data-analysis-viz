### box plot
- 가공하지 않은 자료 그대로를 이용하여 그린 것이 아니라, 자료로부터 얻어낸 통계량인 5가지 요약 수치로 그린다.

- 5가지 요약 수치란 기술통계학에서 자료의 정보를 알려주는 아래의 다섯 가지 수치를 의미한다.
1. 최솟값
2. 제 1사분위 수
3. 제 2사분위 수(), 즉 중앙값
4. 제 3사분위 수()
5. 최댓값

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

%matplotlib inline

# 한글폰트 사용시 그래프에서 마이너스 폰트 깨지는 문제에 대한 대처
mpl.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.DataFrame(np.random.rand(10, 5), columns=['A', 'B', 'C', 'D', 'E'])
```

```python
df
```

```python
# df의 요약된 수치를 출력
df.describe()
```

```python
df.plot.box()
```

```python
color = {'boxes': 'DarkGreen', 'whiskers': 'DarkOrange',
         'medians': 'DarkBlue', 'caps': 'Gray'}
```

```python
# sym='r+': symbol(이상치)는 빨간색 +(플러스 기호)로 표시
df.plot.box(color=color, sym='r+')
```

```python
# vert=False: x축과 y축 반대
# positions=[1, 4, 5, 6, 8]: 박스 플롯이 그려질 좌표 설정(간격 조절)ㅡ
#           현재는 x축과 y축이 바뀌었기 때문에 여기서 지정한 숫자들은 y축의 눈금의 값            
df.plot.box(vert=False, positions=[1, 4, 5, 6, 8])
```

```python
df = pd.DataFrame(np.random.randn(10, 5))

df.head(1)
```

```python
# boxplot(): 박스 플롯 생성, 기본값으로 Grid(그리드)를 그려줌
bp = df.boxplot()
```

```python
# np.random.rand(): 균등 분포(0에서 1미만 사이의 실수만 생성)
# np.random.randn(): 표준 정규 분포(주로 0 근처의 숫자가 나옴, n은 Normal(정규분포)를 의미)
df = pd.DataFrame(np.random.rand(10, 2), columns=['Col1', 'Col2'])

df.head(2)
```

```python
df['X'] = pd.Series(['A', 'A', 'A', 'A', 'A', 'B', 'B', 'B', 'B', 'B'])

df.head(2)
```

```python
# by='X': 'X'열의 값을 기준으로 데이터를 나눔
bp = df.boxplot(by='X')
```

```python
# 난수 생성 엔진의 시작 번호를 고정
np.random.seed(1234)

df_box = pd.DataFrame(np.random.randn(50, 2))

# choice(['A', 'B'], size=50): 'A', 'B' 중 하나를 골라 50개의 행에 할당하는 'g'열 생성
df_box['g'] = np.random.choice(['A', 'B'], size=50)

# 'g'열의 값이 B인 행들 중 1번 인덱스(두번째 열) 값에 3을 더하기
df_box.loc[df_box['g'] == 'B', 1] += 3

# 'g'열의 'A'와 'B'를 기준으로 박스 플롯 생성
bp = df_box.boxplot(by='g')

# 0번 컬럼은 두 그룹이 비슷하게 나타남
# 1번 컬럼은 'B'에 3을 더했기 때문애 'A'보다 더 위쪽에 그려짐
```

```python
# 하나의 창 안에 박스 플롯 생성
bp = df_box.groupby('g').boxplot()
```
